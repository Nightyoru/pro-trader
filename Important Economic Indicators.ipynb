```python
# tools
def AQuarter_compare(q):
    a=q
    a.iloc[:,1] = a.iloc[:,1].pct_change(periods = 4)*100
    return a

def Annual_compare(q):
    a=q
    a.iloc[:,1] = a.iloc[:,1].pct_change(periods = 12)*100
    return a
def Previous_compare(q):
    a=q
    a.iloc[:,1] = a.iloc[:,1].pct_change()*100
    return a
```

## Dashboard Function use to Visualize two Data Series:

```python
def make_dashboard(data1, data2 ,title = None , x = None, y1= None, y2= None, beg =None, end = None, length =None, twin =True,
color1='tab:red',
color2='tab:blue'):
    if x == None:
        x=str(data1.columns[0])
    if y1 == None:
        y1=str(data1.columns[1])
    if y2 == None:
        y2=str(data2.columns[1])
    data1.iloc[:,0] = pd.to_datetime(data1.iloc[:,0])
    data1= data1[data1.iloc[:,1] != '.'] # delete data that does not have number
    data1.iloc[:,1] = pd.to_numeric(data1.iloc[:,1]) #convert to number
    data2.iloc[:,0] = pd.to_datetime(data2.iloc[:,0])
    data2= data2[data2.iloc[:,1] != '.'] # delete data that does not have number
    data2.iloc[:,1] = pd.to_numeric(data2.iloc[:,1]) #convert to number
    data = pd.merge_asof(data1,data2)
    if beg!=None or end!=None:
        data = data.loc[beg:end]
    if length!=None:
        data = data.iloc[-length:]
    if twin ==False:
        data = data.set_index('DATE')
        ax = data.plot(figsize =(20,10),color = 'r''b')
        ax.set_ylabel(y1,fontsize = 15)
        ax.xaxis.label.set_size(15)
        plt.suptitle(title, fontsize= 20)
    else:
        fig,axis1 = plt.subplots( figsize= (20,10))
        plt.title(title, fontsize= 20)
        axis1.set_xlabel(x)
        data = data.set_index('DATE')
        axis1.set_ylabel(y1, color =color1, fontsize = 15)
        axis1.plot(data.iloc[:,0], color = color1)
        
        axis2 = axis1.twinx()
        axis2.set_ylabel(y2, color = color2, fontsize = 15)
        axis2.plot(data.iloc[:,1], color = color2)
    
                  
    
    a = data.iloc[:,0].corr(data.iloc[:,1])
    a = str(a)
    print('The correlation is ' + a)
```

```python
make_dashboard(LIBOR, TenTbond, length= 365*2, y1= 'Rate', title = 'Shorterm Long term Debt-Rate', twin = False)
```

```python
# single data dashboard
def singledata_dashboard(data, y= None ,title = None, start = None, end = None):
    data.iloc[:,0] = pd.to_datetime(data.iloc[:,0])
    data = data.set_index('DATE')
    ax = data.loc[start:end].plot(figsize = 15)
    ax.set_ylabel(y, fontsize = 15)
    ax.xaxis.label.set_size(15)
    plt.suptitle(title, fontsize= 20)
```
